```python
import cv2
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from matplotlib import pyplot as plt
```

```python
import os
```

```python
imgloc=[]
label=[]
for i in range(0,121):
    for dirname, _, filenames in os.walk('C:/Users/AHM ISRAFIL/Documents/ANACONDA/OCR Bangla/data/train/'+str(i)):
        for filename in filenames:
            imgloc.append((os.path.join(dirname, filename)).replace("'", ""))
            label.append(i)
```

```python
img=[]
for i in range(0, len(imgloc)):
    img1 = cv2.imread(imgloc[i],1)
    img2 = np.array(img1)
    retValue, img2 = cv2.threshold(img2, 55, 255, cv2.THRESH_BINARY)
    img2 = cv2.resize(img2,(56,56))
    img.append(cv2.cvtColor(img2,cv2.COLOR_BGR2RGB))
```

```python
plt.figure()
f, axarr = plt.subplots(5, 10) 

for i in range(5):
    for j in range(10):
        index = (10 * i) + j
        if index < len(img):
            axarr[i][j].imshow(img[index])
        else:
            # If index exceeds the number of images, break out of the loop
            break
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(img, label, test_size=0.25, random_state=11)
print (y_train)
```

```python
import pandas as pd

# Read the CSV file containing the mapping
mapping_df = pd.read_csv('C:/Users/AHM ISRAFIL/Documents/ANACONDA/OCR Bangla/data/metaDataCSV.csv')

# Create a dictionary from the mapping dataframe
mapping_dict = dict(zip(mapping_df['Label'], mapping_df['CharName']))

# Replace numbers in y_train and y_test with characters
y_train_chars = [mapping_dict[num] for num in y_train]
y_test_chars = [mapping_dict[num] for num in y_test]

# Print the characters corresponding to the numbers
print(y_train_chars)
print(y_test_chars)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
```

```python
# Data Preprocessing
# Convert images to 1D vectors
x_train_flat = [img.flatten() for img in x_train]
x_test_flat = [img.flatten() for img in x_test]
```

```python
# Normalize pixel values
x_train_normalized = np.array(x_train_flat) / 255.0
x_test_normalized = np.array(x_test_flat) / 255.0
```

```python
# Convert labels to one-hot encoded vectors
num_classes = len(mapping_dict)
y_train_onehot = pd.get_dummies(y_train_chars).values
y_test_onehot = pd.get_dummies(y_test_chars).values
```

```python
# Reshape the flattened images to match the expected LSTM input shape
num_timesteps = 28  # Assuming the images were resized to 56x56 pixels
x_train_reshaped = x_train_normalized.reshape(-1, num_timesteps, num_timesteps)
x_test_reshaped = x_test_normalized.reshape(-1, num_timesteps, num_timesteps)
```

```python
from tensorflow.keras.layers import Dense, Dropout
# Define LSTM Model
model = Sequential()
model.add(LSTM(64, input_shape=(1,2548),activation="relu",return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(32,activation="sigmoid"))
model.add(Dropout(0.2))
#model.add(LSTM(100,return_sequences=True))
#model.add(Dropout(0.2))
#model.add(LSTM(50))
#model.add(Dropout(0.2))
model.add(Dense(3, activation='sigmoid'))
model.compile(loss = 'categorical_crossentropy', optimizer = "adam", metrics = ['accuracy'])
model.summary()
```

```python
# Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Train the model
model.fit(x_train_reshaped, y_train_onehot, batch_size=64, epochs=10, validation_data=(x_test_reshaped, y_test_onehot))
```
